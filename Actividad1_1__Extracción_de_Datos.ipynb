```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

**Carga de archivo**



```python
df= pd.read_excel('Detalle precios y productos fabricados 2022.xlsx')
df
```

```python
df.info()
```

```python
df.head()
```

**Filtro 1:** Por objeto

Se filtran los registros en donde los valores de la columna cant son igual a 1

```python
filtro1=df[df['CANT'] == 1]
filtro1
```

```python
filtro1.to_csv("Por_objeto.csv")
from google.colab import files
files.download("Por_objeto.csv")
```

**Filtro 2:** Filas consegutivas

Se filtran filas consecutivas

```python
filtro2= df.iloc[2:20,: ]
filtro2
```

```python
filtro2.to_csv("filas_consecutivas.csv")
from google.colab import files
files.download("filas_consecutivas.csv")
```

**Filtro 3:** Columnas no consecutivas

Se filtran columnas que no son consecutivas



```python
filtro3=df.iloc[ : , [0,2,6,7]]
filtro3
```

```python
filtro3.to_csv("columnas_no_consecutivas.csv")
from google.colab import files
files.download("columnas_no_consecutivas.csv")
```

**Filtro 4:** Filas vs Columnas

Se define un nuevo indice, el cual será la columna 3

```python
df1= pd.read_excel('Detalle precios y productos fabricados 2022.xlsx',index_col =3)
df1
```

**Filtro 7:** Comparacion

Se filtran los valores cuantitativos de una columna

```python
filtro7=df[df["CANT"] < 2]
filtro7
```

```python
filtro7.to_csv("comparacion.csv")
from google.colab import files
files.download("comparacion.csv")
```

**Filtro 8:** Conectores logicos

Se filtran los registros que tengan un valor menor a 2 en el almacén y durante el día 05/01/2022

```python
filtro8=df[(df["CANT"] < 2) & (df["FECHA_DOC"] == "05/01/2022 00:00")]
filtro8
```

```python
filtro8.to_csv("conectores_logisticos.csv")
from google.colab import files
files.download("conectores_logisticos.csv")
```

**Filtro 9:** Not


Al igual que en el filtro anterior se filtran los registros que tengan un valor menor a 2 en el almacén y durante el día 05/01/2022

```python
filtro9=df[~(df["CANT"] >= 2) & ~(df["FECHA_DOC"] == "05/01/2022 00:00")]
filtro9
```

```python
filtro9.to_csv("not.csv")
from google.colab import files
files.download("not.csv")
```
